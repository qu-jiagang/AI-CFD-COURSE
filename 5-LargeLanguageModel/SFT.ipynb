# 介绍

在大语言模型（LLM, Large Language Model）的领域中，**SFT** 是 **Supervised Fine-Tuning（监督微调）** 的简称。它是对预训练模型进行进一步训练的一种方法，目的是让模型更适应特定任务或领域的需求。

SFT 是在预训练语言模型（例如 GPT 系列、BERT）之上的一个精调过程。预训练模型通常通过大规模的无监督学习（比如语言建模任务）学习到通用的语言表示能力，但这些能力可能不足以胜任特定的任务或领域。这时，通过 SFT，可以利用标注好的高质量数据对模型进行微调，使其在特定任务上的表现更好。
